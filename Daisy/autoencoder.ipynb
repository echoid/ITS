```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Model
```

```python
fd_data = pd.read_csv("modified_adult_train.csv")
```

```python
# input 
A = fd_data["education-num"]

# output
B = fd_data["capital-gain"]
```

## Follow the paper construction, the autoencoder does not work

```python
def autoencoder(kernal_size):
    data = keras.Input(shape=(None,1))
    encoded = layers.Conv1D(1, kernal_size, strides=2)(data)
    decoded = layers.Conv1DTranspose(1,kernal_size,2)(encoded)

    encoder = keras.Model(data, encoded)
    autoencoder = keras.Model(data, decoded)

    autoencoder.compile(optimizer='adam', loss=tf.keras.losses.MeanAbsoluteError())

    return autoencoder
```

```python
auto_encoder = autoencoder(5)
auto_encoder.fit(
  A, B,epochs=5000,batch_size=256)
```

## Use Kernal size == 1, not working

```python
auto_encoder = autoencoder(1)
auto_encoder.fit(
  A, B,epochs=5000,batch_size=256)
```
